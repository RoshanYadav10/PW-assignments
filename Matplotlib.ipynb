```python
# # Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
# # Matplotlib.
# Matplotlib is a popular and powerful Python library used for creating static, interactive, and animated visualizations in various formats.
# Five plots that can be plotted using the Pyplot module of Matplotlib are:
# Line plot: A basic plot that displays data points connected by straight lines, useful for visualizing trends and patterns over continuous data.

# Bar plot: A plot that uses rectangular bars to represent data, where the length of each bar corresponds to the value of the data. Bar plots are 
# often used for comparing different categories or groups.

# Scatter plot: A plot that displays individual data points as dots on a two-dimensional plane. It is useful for visualizing the distribution 
# and relationship between two numerical variables.

# Histogram: A plot that displays the distribution of a single numerical variable by dividing the data into bins and counting the number of 
# data points in each bin.

# Pie chart: A circular chart that represents the proportion of each category relative to the whole. It is useful for displaying the relative 
# sizes of different categories as parts of a whole.
```

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# Scatter plot: A plot that displays individual data points as dots on a two-dimensional plane. It is useful for visualizing the distribution 
# and relationship between two numerical variables.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("xData")
plt.ylabel("yData")
plt.title("Heading")
plt.scatter(x,y)
```

```python
# # Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# The subplot() function is used to create multiple subplots within a single figure. It allows us to compare different plots side by side or 
# create complex visualizations with different subplots representing different aspects of the data.
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])


x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1', color='red')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='blue')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='green')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='orange')

plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# Bar plot: A plot that uses rectangular bars to represent data, where the length of each bar corresponds to the value of the data. Bar plots are 
# often used for comparing different categories or groups.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset.
# The key components of a box plot are:

# Median (Q2): The middle value of the data. 50% of the data falls below and 50% above the median.
# Quartiles (Q1 and Q3): The 25th and 75th percentiles of the data, respectively. Q1 represents the lower quartile (25% of data below),
#  and Q3 represents the upper quartile (75% of data below).
# Interquartile Range (IQR): The range between the first quartile (Q1) and the third quartile (Q3) (IQR = Q3 - Q1).
# Whiskers: Lines extending from the box that indicate the minimum and maximum data points within a defined range from the quartiles.
# Outliers: Data points that fall beyond the whiskers and are considered as potential extreme values.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.boxplot(data)
plt.show()
```
