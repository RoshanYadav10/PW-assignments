```python
# Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
# use each type of test.

# Both t-tests and z-tests are statistical hypothesis tests used to make inferences about population parameters based on sample data. However, 
# they are used under different circumstances and assumptions, primarily due to differences in the sample size and our knowledge about population 
# parameters.

# **T-Test:**
# A t-test is used when the sample size is relatively small and the population standard deviation is unknown. The t-test takes into account the 
# variability of the sample data to estimate the population parameters. There are different variations of t-tests, such as the independent samples 
# t-test for comparing two independent groups and the paired samples t-test for comparing two related groups.

# **Example Scenario for T-Test:**
# Suppose you are a researcher studying the effects of a new drug on blood pressure. You have two groups: a control group and a treatment group. 
# You measure the blood pressure of participants in both groups before and after administering the treatment. To determine if the treatment has a 
# significant effect on blood pressure, you would use a paired samples t-test to compare the mean blood pressure before and after treatment within 
# each group.

# **Z-Test:**
# A z-test is used when the sample size is large (typically n > 30) and the population standard deviation is known or can be reasonably assumed. 
# The z-test relies on the assumption of a normal distribution of the sample data. It's often used for hypothesis testing involving population 
# proportions or means when the sample size is sufficiently large.

# **Example Scenario for Z-Test:**
# Imagine you work for a marketing team and want to test if a new email campaign leads to a significantly higher click-through rate compared to 
# the old campaign. You randomly select a large sample of email recipients and record whether they clicked the link in the email or not. Since you 
# have a large sample and you're interested in comparing proportions (click-through rates), you can use a z-test to determine if there's a 
# statistically significant difference in click-through rates between the two campaigns.

# In summary:
# - Use a t-test when the sample size is small, and the population standard deviation is unknown.
# - Use a z-test when the sample size is large, and either the population standard deviation is known or can be assumed, especially for proportion-
# related hypotheses.

```

```python
# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
# each type of error.
# Type 1 error:
# When we reject Null hypothesis when in reality it is true
# Example Scenario for Type 1 Error:
# Suppose a pharmaceutical company is testing a new drug for effectiveness in reducing blood pressure. The null hypothesis (H0) states that the 
# drug has no effect on blood pressure. However, due to random variability in the sample data, the statistical test leads to rejecting the null 
# hypothesis and concluding that the drug is effective in reducing blood pressure, even though it actually isn't. This would be a Type 1 error.

# Type 2 Error:
# When we retain Null hypothesis when in reality it is false
# Example scenario for Type 2 error:
# Imagine a quality control scenario where you're testing a manufacturing process to ensure that a certain component meets specific standards. The
#  null hypothesis (H0) states that the process is producing components according to the standards. However, due to sampling variability or 
#  measurement errors, the statistical test fails to reject the null hypothesis, even though the process is actually producing substandard 
#  components. This would be a Type 2 error.
```

```python
# Q4: Explain Bayes's theorem with an example.
# Bayes's Theorem is a fundamental concept in probability theory and statistics that allows us to update our beliefs or probabilities about an 
# event based on new evidence. It provides a way to calculate the probability of a hypothesis (event) given some observed evidence. The theorem
#  is named after the Reverend Thomas Bayes, an 18th-century mathematician and theologian.

# The formula for Bayes's Theorem is as follows:

#  P(A|B) = {P(B|A) * P(A)}/{P(B)} 

# Where:
# -  P(A|B) is the probability of event A occurring given that event B has occurred.
# -  P(B|A)  is the probability of event B occurring given that event A has occurred.
# -  P(A)  is the prior probability of event A.
# -  P(B)  is the probability of event B.

# Example Scenario: Medical Test
# Let's use a classic example involving a medical test to illustrate Bayes's Theorem. Suppose you're concerned about having a rare disease, which
#  we'll call Disease X. The overall prevalence of Disease X in the population is very low, say 0.1%.

# You decide to take a diagnostic test for Disease X, which isn't perfect and sometimes gives false positive results. The test has the following 
# properties:
# - Sensitivity: The probability of testing positive given that you actually have the disease is 95%.
# - Specificity: The probability of testing negative given that you don't have the disease is 90%.

# Now, you've taken the test and received a positive result (Event B). What you want to know is the probability that you actually have the disease
#  (Event A), given this positive test result.

# Using Bayes's Theorem:
# - P(A)  (prior probability of having the disease) = 0.001 (0.1% prevalence)
# -  P(B|A)  (probability of testing positive given you have the disease) = 0.95 (sensitivity)
# -  P(B|- A)  (probability of testing positive given you don't have the disease) = 1 - 0.90 = 0.10 (1 - specificity)
# -  P(-A)  (probability of not having the disease) = 1 - 0.001 = 0.999

# Now plug these values into the formula:
#  P(A|B) = {0.95*0.001}/{0.95*0.001 + 0.10 * 0.999} approx 0.0094

# So, even though you received a positive test result, the probability that you actually have the disease is only about 0.94%. This demonstrates 
# the importance of considering both the sensitivity and specificity of a test along with the prevalence of the condition when interpreting test
#  results.
```

```python
# Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
# event's probability and new evidence. Provide a sample problem and solution.
# Sample Problem:
# Suppose you work for an online retailer and you want to estimate the probability that a customer will make a purchase on your website. You have historical data that shows that, on average, 20% of visitors to your website make a purchase (prior probability). You've also conducted a study and found that customers who spend more than 5 minutes browsing the website are more likely to make a purchase. Specifically, among those who spend more than 5 minutes on the website, 40% make a purchase (conditional probability). For those who spend 5 minutes or less on the website, only 10% make a purchase.

# Now, a new visitor has just spent more than 5 minutes on your website. What is the probability that this visitor will make a purchase?

# **Solution:**
# Let's use Bayes's Theorem to calculate the probability of making a purchase given the new evidence (visitor spent more than 5 minutes on the website).

# We'll define the events:
# - Event A: The visitor makes a purchase.
# - Event B: The visitor spent more than 5 minutes on the website.

# Given information:
# -  P(A) : Prior probability of making a purchase = 0.20
# -  P(B|A) : Probability of spending more than 5 minutes given making a purchase = 0.40
# -  P(B|- A) : Probability of spending more than 5 minutes given not making a purchase = 1.00 (everyone who spends time browses more than 5 minutes)
# -  P(-A): Probability of not making a purchase = 1 -  P(A)  = 0.80

# Now we can use Bayes's Theorem:

# P(A|B) = {P(B|A)*P(A)}/{P(B|A)*P(A) + P(B|-A)*P(-A)} 

# Substitute the values:
# P(A|B) = {0.40*0.20}/{0.40*0.20 + 1.00*0.80} approx 0.20 

# So, based on the new evidence that the visitor spent more than 5 minutes on the website, the probability that this visitor will make a purchase 
# is approximately 0.20, or 20%.

# Bayes's Theorem allows us to update our probability estimate based on new evidence, helping us make more informed decisions or predictions.
```

```python
# # Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
# # of 5. Interpret the results.

# To calculate the 95% confidence interval for a sample with a mean of 50 and a standard deviation of 5, we can use the formula:

# Confidence Interval = Mean ± (Critical Value) * (Standard Deviation / √(Sample Size))

# Since you didn't provide the sample size or the critical value, I'll assume that the sample size is reasonably large and we can use the Z-score
#  corresponding to a 95% confidence level, which is approximately 1.96.

# Given:
# Mean (μ) = 50
# Standard Deviation (σ) = 5
# Sample Size (n) = Let's assume it's a reasonably large value

# Critical Value (Z) for a 95% confidence level ≈ 1.96

# Confidence Interval = 50 ± (1.96) * (5 / √(n))

# Now, let's interpret the results:

# A 95% confidence interval means that if we were to take multiple samples from the same population and calculate their means, approximately 95% 
# of those sample means would fall within this calculated interval.

# For the given data, with a mean of 50 and a standard deviation of 5, the 95% confidence interval provides a range of values within which we can
#  be reasonably confident that the true population mean lies. In other words, we're 95% confident that the true mean of the population falls within this interval.

# Interpreting the confidence interval:
# - Lower bound: 50 - (1.96) * (5 / √(n))
# - Upper bound: 50 + (1.96) * (5 / √(n))

# Keep in mind that the larger the sample size, the narrower the confidence interval becomes, reflecting increased confidence in the accuracy of 
# our estimate of the true population mean. Conversely, a smaller sample size will result in a wider confidence interval and less confidence in
#  the estimate.
```

```python
# # Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
# # Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

# The margin of error in a confidence interval is a measure of the uncertainty or range within which the true population parameter (such as the 
# mean or proportion) is likely to fall. It represents the maximum amount by which the sample estimate (such as the sample mean or sample 
# proportion) might differ from the true population parameter while still maintaining a certain level of confidence.

# In mathematical terms, the margin of error (MOE) can be calculated as:

# Margin of Error = Critical Value * (Standard Deviation / √(Sample Size))

# Where the critical value is typically derived from the desired confidence level and the distribution of the data (e.g., Z-score for a normal 
# distribution).

# The sample size directly affects the margin of error. Specifically, as the sample size increases, the margin of error decreases. This is because 
# larger sample sizes provide more information about the population, leading to more precise estimates. In other words, with a larger sample size, 
# the sample estimate tends to be closer to the true population parameter, resulting in a narrower range of possible values for the parameter's 
# estimate.

# Here's an example to illustrate the effect of sample size on the margin of error:

# Scenario: Let's say a polling organization wants to estimate the proportion of people in a city who support a particular political candidate.
#  They conduct two separate surveys with different sample sizes.

# Survey 1:
# - Sample Size (n1) = 200
# - Estimated Proportion (p1) = 0.60 (60%)
# - Standard Deviation = √(p1 * (1 - p1)) ≈ 0.49 (estimated based on the sample proportion)

# Survey 2:
# - Sample Size (n2) = 1000 (larger sample)
# - Estimated Proportion (p2) = 0.60 (same as in Survey 1)
# - Standard Deviation = √(p2 * (1 - p2)) ≈ 0.49

# Assuming a 95% confidence level and using the Z-score of approximately 1.96 for a normal distribution:

# Margin of Error for Survey 1 = 1.96 * (0.49 / √(200)) ≈ 0.069 (6.9%)
# Margin of Error for Survey 2 = 1.96 * (0.49 / √(1000)) ≈ 0.031 (3.1%)

# In this example, the larger sample size (Survey 2) results in a smaller margin of error compared to the smaller sample size (Survey 1). This 
# means that the estimate from Survey 2 is more precise and has a narrower range of possible values around the true population proportion.
```

```python
# Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
# standard deviation of 5. Interpret the results.

def calculate_z_score(data_point, population_mean, population_std_dev):
    z_score = (data_point - population_mean) / population_std_dev
    return z_score

data_point = 75
population_mean = 70
population_std_dev = 5

z_score = calculate_z_score(data_point, population_mean, population_std_dev)
print("Z-score:", z_score)

# Interpretation of the result:
# A z-score of 1 indicates that the data point with a value of 75 is 1 standard deviation above the population mean of 70. In other words, it 
# is slightly above the average value within the population.
```

```python
# Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
# of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
# significantly effective at a 95% confidence level using a t-test.


import scipy.stats as stats

# Given data
sample_mean = 6
population_mean = 0  # Null hypothesis: no effect
sample_std_dev = 2.5
sample_size = 50
confidence_level = 0.95

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

# Degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical t-value for a one-tailed test (upper tail)
alpha = 1 - confidence_level
critical_t_value = stats.t.ppf(1 - alpha, df=degrees_of_freedom)

# Compare t-statistic with critical t-value
if t_statistic > critical_t_value:
    conclusion = "Reject the null hypothesis. The drug is significantly effective."
else:
    conclusion = "Fail to reject the null hypothesis. There is no significant evidence of drug effectiveness."

# Print results
print("Calculated t-statistic:", t_statistic)
print("Critical t-value:", critical_t_value)
print("Conclusion:", conclusion)

```

```python
# Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
# confidence interval for the true proportion of people who are satisfied with their job.

import math

sample_proportion = 0.65
confidence_level = 0.95
z_score = 1.96
sample_size = 500

standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

margin_of_error = z_score * standard_error

confidence_interval_lower = sample_proportion - margin_of_error
confidence_interval_upper = sample_proportion + margin_of_error

print("95% Confidence Interval:", (confidence_interval_lower, confidence_interval_upper))
```

```python
# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
# Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
# with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
# significant difference in student performance using a t-test with a significance level of 0.01.

import scipy.stats as stats

# Given data for Sample A
mean_a = 85
std_dev_a = 6
sample_size_a =20

# Given data for Sample B
mean_b = 82
std_dev_b = 5
sample_size_b =20

# Significance level
alpha = 0.01

# Calculate the pooled standard deviation
pooled_std_dev = math.sqrt(((std_dev_a ** 2) * (sample_size_a - 1) + (std_dev_b ** 2) * (sample_size_b - 1)) / (sample_size_a + sample_size_b - 2))

# Calculate the t-statistic
t_statistic = (mean_a - mean_b) / (pooled_std_dev * math.sqrt(1/sample_size_a + 1/sample_size_b))

# Degrees of freedom
degrees_of_freedom = sample_size_a + sample_size_b - 2

# Calculate the critical t-value
critical_t_value = stats.t.ppf(1 - alpha/2, df=degrees_of_freedom)

# Compare t-statistic with critical t-value
if abs(t_statistic) > critical_t_value:
    conclusion = "Reject the null hypothesis. There is a significant difference in student performance."
else:
    conclusion = "Fail to reject the null hypothesis. There is no significant difference in student performance."

# Print results
print("Calculated t-statistic:", t_statistic)
print("Critical t-value:", critical_t_value)
print("Conclusion:", conclusion)
```

```python
# Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
# of 65. Calculate the 90% confidence interval for the true population mean.

sample_mean = 65
confidence_level = 0.90
z_score = 1.645
standard_deviation = 8
sample_size = 50

standard_error = standard_deviation / math.sqrt(sample_size)

margin_of_error = z_score * standard_error

confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print("90% Confidence Interval:", (confidence_interval_lower, confidence_interval_upper))
```
