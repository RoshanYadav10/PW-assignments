```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Estimation in statistics refers to the process of making educated guesses or predictions about population parameters based on sample data. 
Estimation involves using sample statistics to estimate the corresponding population parameters. Two common types of estimation are point 
estimation and interval estimation.

Point Estimate:
A point estimate is a single value that serves as the "best guess" or estimate of a population parameter. It is obtained by using a sample
 statistic as a substitute for the unknown population parameter.

Interval Estimate (Confidence Interval):
An interval estimate, also known as a confidence interval, provides a range of values within which the population parameter is likely to fall
 with a certain level of confidence.
```

```python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.

import scipy.stats as stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    """
    Conducts a two-sample t-test on the difference between two population means.
    
    :param sample1: List or array of data for sample 1.
    :param sample2: List or array of data for sample 2.
    :param alpha: Significance level for the test (default is 0.05).
    :return: A tuple containing the t-statistic, p-value, and whether the null hypothesis is rejected.
    """
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)
    reject_null = p_value < alpha
    return t_statistic, p_value, reject_null

# Example data for two independent samples
sample1 = [68, 70, 72, 65, 67, 69, 71, 73, 66, 68]
sample2 = [60, 62, 61, 59, 64, 63, 65, 61, 63, 62]

# Conduct the two-sample t-test
t_statistic, p_value, reject_null = two_sample_t_test(sample1, sample2)

# Print the results
print("Two-Sample T-Test Results:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
print("Reject Null Hypothesis:", reject_null)
```

```python
# Q8. Define p-value and explain its significance in hypothesis testing.

# The p-value, short for "probability value," is a fundamental concept in hypothesis testing. It quantifies the strength of evidence against the 
# null hypothesis (H0) based on the observed sample data. In other words, the p-value helps you determine whether the observed data provides enough 
# evidence to either reject or fail to reject the null hypothesis

# Significance of p-value in Hypothesis Testing:
# The p-value plays a crucial role in hypothesis testing for the following reasons:

# Decision Rule: In hypothesis testing, you compare the p-value to a predetermined significance level (alpha) to make a decision about the null 
# hypothesis. If the p-value is less than or equal to alpha, you reject the null hypothesis. If the p-value is greater than alpha, you fail to 
# reject the null hypothesis.

# Measuring Evidence: The p-value provides a measure of the strength of evidence against the null hypothesis. A small p-value indicates that the 
# observed data is unlikely to have occurred if the null hypothesis were true, suggesting that there is strong evidence against the null hypothesis.

# Interpretation: A low p-value suggests that the observed data is inconsistent with the null hypothesis. It suggests that the data provides 
# support for the alternative hypothesis or the presence of a significant effect or relationship.

```

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.

import numpy as np
import scipy.stats as stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    """
    Conducts a two-sample t-test for independent samples assuming equal variances.
    
    :param sample1: List or array of data for sample 1.
    :param sample2: List or array of data for sample 2.
    :param alpha: Significance level for the test (default is 0.05).
    :return: A tuple containing the t-statistic, p-value, and whether the null hypothesis is rejected.
    """
    # Calculate sample statistics
    n1 = len(sample1)
    n2 = len(sample2)
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)
    std2 = np.std(sample2, ddof=1)
    
    # Calculate pooled standard deviation
    pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))
    
    # Calculate t-statistic
    t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(1/n1 + 1/n2))
    
    # Calculate degrees of freedom
    df = n1 + n2 - 2
    
    # Calculate p-value
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    # Determine whether to reject the null hypothesis
    reject_null = p_value < alpha
    
    return t_statistic, p_value, reject_null

# Example random samples
sample1 = np.array([23, 25, 28, 31, 26])
sample2 = np.array([20, 22, 24, 27, 25])

# Conduct the two-sample t-test
t_statistic, p_value, reject_null = two_sample_t_test(sample1, sample2)

# Print the results
print("Two-Sample T-Test Results:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
print("Reject Null Hypothesis:", reject_null)
```

```python
# # Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
# # clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
# # standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

# To test the hypothesis about the decrease in blood pressure, we can use a one-sample t-test since we are comparing the sample mean to a 
# hypothesized population mean. The null hypothesis (H0) assumes that the drug does not decrease blood pressure by 10 mmHg, while the alternative
#  hypothesis (H1) suggests that the drug does decrease blood pressure by 10 mmHg.

# The hypotheses are:
# - Null Hypothesis (H0): The drug does not decrease blood pressure by 10 mmHg. μ = 10
# - Alternative Hypothesis (H1): The drug does decrease blood pressure by 10 mmHg. μ < 10

# Given information:
# - Sample Mean (x̄) = 8 mmHg
# - Sample Standard Deviation (s) = 3 mmHg
# - Sample Size (n) = 100
# - Significance Level (α) = 0.05

# First, we need to calculate the t-score:
# t = (x̄ - μ) / (s / √n)
# t = (8 - 10) / (3 / √100)
# t = -2 / 0.3
# t ≈ -6.67

# Next, we need to find the critical t-value for a one-tailed test with 99 degrees of freedom (n - 1) at a significance level of 0.05. Using a
#  t-distribution table or calculator, the critical t-value is approximately -1.660.

# Since the calculated t-score (-6.67) is much lower (more negative) than the critical t-value (-1.660), we can reject the null hypothesis.

# Based on the data from the clinical trial and the significance level of 0.05, we have found evidence to suggest that the new drug's effect on blood pressure may not be accurately described 
# as a decrease of 10 mmHg. The sample data indicates a decrease of approximately 8 mmHg, but we cannot definitively conclude that the effect is 
# less than 10 mmHg based solely on this study. Further research may be needed to determine the true effect size.
```
